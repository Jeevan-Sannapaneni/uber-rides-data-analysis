```python
import pandas as pd
df = pd.read_csv('My Uber Drives - 2016.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['PURPOSE*'] = df['PURPOSE*'].fillna('Not Specified')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(data=df,x='CATEGORY*')
plt.title("Ride Category Distribution")
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, y='PURPOSE*',order=df['PURPOSE*'].value_counts().index)
plt.title("Trip Purpose Distribution")
```

```python
sns.histplot(df['MILES*'],bins=40)
plt.title("Trip Distance Distribution")
```
